```python
nltk.download('all')

caminho_arquivo_csv = '/content/beer_reviews.csv'

# Lê o arquivo CSV em um DataFrame
df = pd.read_csv(caminho_arquivo_csv)

# Exibe as primeiras linhas do DataFrame
df.head()
```

```python
# Criação do analisador de sentimentos
analyzer = SentimentIntensityAnalyzer()

# Função para classificar sentimentos em um DataFrame
def classify_sentiment(description):
    sentiment_score = analyzer.polarity_scores(description)
    return sentiment_score['compound']

# Aplicar a função "classify_sentiment" à coluna "Description" do DataFrame e armazenar o resultado na coluna "Description"
df['Sentiment(Recomendation)'] = df['Description'].apply(classify_sentiment)
```

```python
X_train, X_test = train_test_split(df, test_size = 0.30, random_state = 42)

print(X_train.shape)
print(X_test.shape)

X_train
```
