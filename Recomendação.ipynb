Atribuindo um id único a cada avaliador da coluna "review_profilename"

```python
# Suponha que a coluna com os nomes dos avaliadores é chamada 'Nome_Avaliador'
avaliadores = df['review_profilename'].unique()
avaliador_id_map = {nome: i for i, nome in enumerate(avaliadores, 1)}

df['ID_Avaliador'] = df['review_profilename'].map(avaliador_id_map)

df

```

```python
df.columns
```

Descartando algumas colunas que não serão úteis

```python
df = df[['beer_beerid', 'beer_name', 'beer_abv', 'beer_style', 'ID_Avaliador', 'review_overall']]
df
```

Alterando o nome das colunas

```python
columns = {'beer_beerid':'id_cerveja', 'beer_name':'nome_cerveja', 'beer_abv':'teor_alcoolico', 'beer_style':'tipo_cerveja', 'review_overall':'nota'}
df.rename(columns = columns, inplace=True)
df
```

Quantos usuários avaliaram pelo menos 50 cervejas?


```python
y = df['ID_Avaliador'].value_counts() >= 50
y[y == True]
```

De 33388 usuários, 4732 avaliaram pelo menos 50 cervejas! Queremos utilizar apenas esses usuários.

```python
w = y[y].index.tolist() #Lista dos ID's dos avaliadores que avaliaram pelo menos 50 cervejas
av_50 = df[df['ID_Avaliador'].isin(w)]
av_50
```
